#1. Qual carro tem a maior quantidade de cavalos (hp)?

```python
from google.colab import files
uploaded = files.upload()
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('mtcars.csv')
carro_mais_potente = df[df['hp'] == df['hp'].max()]
plt.bar(df['model'], df['hp'], color='lightblue')
plt.xticks(rotation=90)
plt.xlabel('Modelos de Carros')
plt.ylabel('Cavalos de Força (hp)')
plt.title('Potência dos Carros (hp)')
plt.legend()
plt.show()
```

#2. Qual a distribuição dos carros de acordo com o número de marchas (gear)?

```python
df_gear = df['gear'].value_counts()
df_gear.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'cyan', 'lightgrey'])
plt.title('Distribuição de Carros por Número de Marchas')
plt.ylabel('')
plt.show()
```

#3. Qual é a média de milhas por galão (mpg) para cada número de marchas (gear)?

```python
df_grouped_gear_mpg = df.groupby('gear')['mpg'].mean()
df_grouped_gear_mpg.plot(kind='bar', color=['purple', 'grey', 'forestgreen'])
plt.title('Média de Milhas por Galão por Número de Marchas')
plt.xlabel('Número de Marchas (gear)')
plt.ylabel('Média de Milhas por Galão (mpg)')
plt.xticks(rotation=0)
plt.show()
```

#4. Quais São os carros com maior quantidade de Marchas (gears)?

```python
max_marchas = df['gear'].max()
carros_max_marchas = df[df['gear'] == max_marchas]
print(carros_max_marchas[['model', 'gear']])
plt.barh(df['model'], df['gear'], color='orange')
plt.xlabel('Carros')
plt.ylabel('Número de Marchas')
plt.title('Número de Marchas por Carro')
plt.tight_layout()
plt.show()
```

#5. Qual a relação entre Cavalos (hp) e Milhas por galão (mpg)?

```python
plt.scatter(df['hp'], df['mpg'])
plt.title('Relação entre Potência (hp) e Milhagem (mpg)')
plt.xlabel('Cavalos (hp)')
plt.ylabel('Milhas por galão (mpg)')
plt.show()
```
